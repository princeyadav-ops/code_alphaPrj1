```python
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("netflix_titles.csv")
df.shape
df.head()
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace= True)
df.shape
```

```python
df['type'] = df['type'].dropna()
type_counts = df['type'].value_counts()
type_counts.plot(kind = 'bar', color= ['red', 'purple'][:len(type_counts)])
plt.title("Movies vs tv shows")
plt.xlabel("Type")
plt.ylabel("count")
plt.grid(axis = 'y')
plt.show()
```

```python
df['date_added'] = pd.to_datetime(df['date_added'], errors= 'coerce')
df['year_added'] = df['date_added'].dt.year

df['year_added'].value_counts().sort_index().plot(kind= 'line', color = 'green')
plt.title('content over time')

plt.grid(True)
plt.show()
```

```python
df['country'].value_counts().head(15).plot(kind= 'bar', color = 'orange')
plt.title('Top 15 countries with most contents')
plt.show()
```

```python


rating_counts = df['rating'].value_counts()

plt.figure(figsize=(12,8))

plt.pie(
    rating_counts,
    labels=rating_counts.index,
    autopct=lambda p: f'{p:.1f}%' if p > 3 else '',
    startangle=140
)

plt.title('Rating Distribution', fontsize=16, fontweight='bold')
plt.tight_layout()
plt.show()
```
